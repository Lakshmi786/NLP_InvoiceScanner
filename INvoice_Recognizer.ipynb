```python
import re

text = extractedInformation
# Split the text into lines
lines = text.split("\n")

# Define the keys to look for in the text
keys = ["BILLTO", "Invoice #", "Task", "Item", "Rate", "Price", "Amount Due (USD)"]

# Create a dictionary to store the key-value pairs
kv_pairs = {}

# Define the current key
current_key = None

# Iterate through the lines
for i, line in enumerate(lines):
    # Remove leading/trailing whitespace from the line
    line = line.strip()

    # Check if the line starts with a key
    for key in keys:
        if line.startswith(key):
            # Set the current key
            current_key = key

            # Check if the line contains a value
            if ":" in line:
                # Extract the value for the key
                value = line.split(":")[-1].strip()

                # Check if the value is a monetary value
                if re.match(r"^\$\d+(,\d+)*(\.\d+)?$", value):
                    # Convert the value to a float
                    value = float(value.replace("$", "").replace(",", ""))

                # Add the key-value pair to the dictionary
                kv_pairs[current_key] = value

            break
    else:
        # Check if the current key is not None
        if current_key is not None:
            # Check if the current key is "BILLTO"
            if current_key == "BILLTO":
                # Add the line to the value for the key
                kv_pairs[current_key] = kv_pairs.get(current_key, "") + line + "\n"

            # Check if the current key is "Item"
            elif current_key == "Item":
                # Add the line to the value for the key
                kv_pairs[current_key] = kv_pairs.get(current_key, "") + line + "\n"

# Print the key-value pairs
print(kv_pairs)
```

```python
import gradio as gr
import pytesseract
from PIL import Image
import re

def extract_text_from_image(image_path):
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image)
    return text

def extract_kvp(text):
    # Define the regular expressions to match the keys
    key_regexes = [
        re.compile(r"^BILLTO"),
        re.compile(r"^Invoice #"),
        re.compile(r"^Task"),
        re.compile(r"^Item"),
        re.compile(r"^Rate"),
        re.compile(r"^Price"),
        re.compile(r"^Amount Due \(USD\)"),
    ]

    # Split the text into lines
    lines = text.split("\n")

    # Iterate over the lines and extract the key-value pairs
    kvp = {}
    key = None
    for line in lines:
        for regex in key_regexes:
            if regex.match(line):
                key = regex.pattern.replace("^", "").strip()
                kvp[key] = ""
                break
        if key:
            kvp[key] += line + "\n"

    return kvp

def extract_info(inputs):
    image_path = inputs["image"]
    text = extract_text_from_image(image_path)
    kvp = extract_kvp(text)
    return kvp

inputs = gr.inputs.Image(shape=(None, None, 3), key="image")
outputs = gr.outputs.Textbox(shape=(None,), key="kvp")

app = gr.Interface(
    extract_info, 
    inputs, 
    outputs, 
    title="Key-Value Pair Extractor from Image", 
    description="This app takes an image as input and extracts the key-value pairs from the text in the image using regular expressions."
)

app.launch()

```
